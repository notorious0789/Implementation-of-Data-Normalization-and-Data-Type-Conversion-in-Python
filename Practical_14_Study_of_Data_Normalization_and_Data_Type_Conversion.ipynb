##Data Normalization

```python
import pandas as pd
import numpy as np

data={
    'Product':['Laptop','Mobile','Tablet','Headphones','Smartwatch','Camera'],
    'Price':[55000,20000,30000,2000,8000,45000],
    'Units_Sold':[25,60,35,80,50,15],
    'Discount':[10,15,12,20,18,8]
}

df=pd.DataFrame(data)
print('Original Dataset')
print(df)
```

```python
#Min-Max Normalization
#For Price
df['Price_MinMax']=(df['Price']-df['Price'].min())/(df['Price'].max()-df['Price'].min())

print("Min-Max Normalized Price")
print(df[['Product','Price','Price_MinMax']])
```

```python
#Normalize Multiple Columns
cols=['Price','Units_Sold','Discount']

df_norm=(df[cols]-df[cols].min())/(df[cols].max()-df[cols].min())

print("Normalized Dataset")
print(df_norm)
```

```python
#Z-score Normailzation
df['Units_Zscore']=(df['Units_Sold']-df['Units_Sold'].mean())/df['Units_Sold'].std()

print("Z-score Normalized Units_Sold")
print(df[['Product','Units_Sold','Units_Zscore']])
```

```python
#Decimal Scaling
df['Price_Scaled']=df['Price']/100000

print("Decimal Scaled Price")
print(df[['Product','Price','Price_Scaled']])
```

## Turning Categorical Variables into quantitative variables

```python
data={
    'Order_ID':[101,102,103,104,105,106],
    'Customer_Gender':['Male','Female','Female','Male','Female','Male'],
    'Payment_Method':['UPI','Credit Card','Debit Card','COD','UPI','Debit Card'],
    'Product_Category':['Electronics','Clothing','Home','Electronics','Beauty','Electronics'],
    'City':['Pune','Mumbai','Delhi','Bangalore','Pune','Hyderabad'],
    'Order_Value':[4500,2300,3200,5200,1800,4100]
}

df=pd.DataFrame(data)
print("Original Dataset")
print(df)
```

```python
#Label Encoding
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['Gender_Label'] = le.fit_transform(df['Customer_Gender'])
df
```

```python
#One Hot Encoding
df_encoded = pd.get_dummies(df, columns=['Payment_Method'])
df_encoded
```

```python
#Encoding Multiple Columns
df_multi=pd.get_dummies(df,columns=['Product_Category','City'])
df_multi
```

```python
#Dummy Encoding (Drop First)
df_dummy=pd.get_dummies(df,columns=['Payment_Method'],drop_first=True)
df_dummy
```
